```python
import pandas as pd
couleurs = ["blue","pink","yellow","purple"]
occurence = [20,15,43,6] # le nbm d'occurences doit etre = au nbr de couleurs
S = pd.Series(occurence, index = couleurs)
print(S) 
```

```python
import pandas as pd
color=["pink", "white", "black", "blue"]
S1=pd.Series([20, 15, 6, 43], index=color)
S2=pd.Series([3, 22, 9, 10], index=color)
print(S1+S2)
```

```python
# CREATING DATA FRAMES FROM A LIST on a crée aec la ligne
import pandas as pd # sympa a utiliser
liste = [["Jack",34,"Paris"],["Thomas",30,"Roma"],["Alexandre",16,"New York"]]
df = pd.DataFrame(liste,columns=["name","age","city"])
print(df)
```

```python
# CREATING DATA FRAMES FROM A DICTIONNARY on a crée avec la colonne
import pandas as pd
dictionnary = {"name":["Jack","Thomas","Alexandre"],"age":[34,30,16],"city":["New York","Paris","London"]}
df1= pd.DataFrame(dictionnary)
print(df1)
```
